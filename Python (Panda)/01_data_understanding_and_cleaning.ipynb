```python
#drop null value
df.dropna(subset=['review_headline', 'review_body'], inplace=True)
```

```python
df['review_date'] = pd.to_datetime(df['review_date'], format='%d-%m-%Y')
```

```python
df.info()
```

```python
import re

def clean_text(text):
    if pd.isna(text):
        return ""
    text = text.lower()
    text = re.sub(r'[^a-z0-9\s]', '', text)
    return text.strip()

df['review_body_clean'] = df['review_body'].apply(clean_text)
df['review_headline_clean'] = df['review_headline'].apply(clean_text)
```

```python
#Memory optimization helpful for 30k rows for group by
cat_cols = ['marketplace', 'vine', 'verified_purchase', 'product_category']
df[cat_cols] = df[cat_cols].astype('category')
```

##Key Findings

Several important observations were identified during the data understanding phase:

* Customer satisfaction appears to be high, with an average rating of 4.33 out of 5.
* Positive reviews account for over 83% of all reviews.
* Most reviewers are verified purchasers, representing 87% of total reviews.
* The dataset contains reviews for 21 products within a single category.
* Data quality is excellent, with almost no missing values after cleaning.
* The review period spans approximately 11 months, enabling monthly trend analysis.
